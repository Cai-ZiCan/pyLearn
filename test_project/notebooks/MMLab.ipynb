```python
## 1.数据清洗，将数据由0，255转为0，1
from osgeo import gdal
import os

def preprocess_labels(label_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for f in os.listdir(label_dir):
        with gdal.Open(os.path.join(label_dir, f)) as img:
             img = img.ReadAsArray()
             # 将 255 变为 1
             img[img == 255] = 1
             # 将数组保存为GeoTIFF格式
             driver = gdal.GetDriverByName('GTiff')
             out_img = driver.Create(os.path.join(output_dir, f), img.shape[1], img.shape[0], 1, gdal.GDT_Byte)
             out_img.GetRasterBand(1).WriteArray(img)
             out_img = None

if __name__ == "__main__":
    label_dir = r"D:\\pyLearn\\test_project\\data\\whu_cd\\2012\\splited_images\\train\\label"
    output_dir = r"D:\\pyLearn\\test_project\\data\\whu_cd\\train\\label\\2012"
    preprocess_labels(label_dir, output_dir)
```
